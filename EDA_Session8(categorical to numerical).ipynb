***One hot encoder***
- One hot means if one is ON another one is OFF
- ON represents with 1,and OFF represents with 0
- For example case status has 2 unique labels
    - Denied
    - Certified
- If we apply one hot encoder on case status column it will create 2 extra columns
     - case_status_Denied
     - case_status_Certified
 

```python
visa_df=pd.read_csv("C:\\Users\\LENOVO\\OneDrive\\DATA SCIENCE\\data files\\Visadataset - Visadataset.csv")
```

**Advantages**
- In machine learning we have one important property is **Independent of variables**
- Machine learning models expects variables are **independent each other**
- One hot encoder creating new columns,the new columns are **independent each other**
- Independent variables means both variables have **90 degrees phase shift**
- 90 degrees phase shift also called as **perpendicular to each other**
- perpendicular to each other means **orthogonal to each other**
- orthogonal to each other means **independent to each other**

***Disadvantages***
- One hot encoder creates new columns equal to number of unique labels
- If number of unique labels are increase new columns also increase
- The more columns leads data is more
- Then data is more the complexity will increase
- The complexity increases the time complexity will increases
- Computation increases more computation power required
- Also time also increase
- The entire process names as **curse of dimensionality**
