```python
unique=visa_df["education_of_employee"].unique()
number=[]
for i in unique:
    con=visa_df["education_of_employee"]==i
    val=len(visa_df[con])
    number.append(val)
    print(f"{i}:{val}")
number
```

```python
education_of_employee_df=pd.DataFrame(zip(unique,number),columns=["education_of_employee","Count"])
education_of_employee_df
```
