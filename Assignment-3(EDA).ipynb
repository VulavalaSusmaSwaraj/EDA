***np.where***

- It is only applicable for binary values
- There are 3 arguments
   - condition
   - True
   - False

```python
visa_df=pd.read_csv("C:\\Users\\LENOVO\\OneDrive\\DATA SCIENCE\\data files\\Visadataset - Visadataset.csv")
```

```python
con=visa_df["case_status"]=="Certified"
visa_df["case_status"]=np.where(con,0,1)
visa_df
```
