***Normalization***
- Read the data again
- step-1:take prevailing_wage column:**visa_df["prevailing_wage"]**
- step-2:calculate the min value prevailing wage:min=**visa_df["prevailing_wage"].min**
- step-3:calculate the max value prevailing wage:man=**visa_df["prevailing_wage"].max**
- step-4:calcualte the Nr:step-1-step-2=**visa_df["prevailing_wage"]min**
- step-5:DR=step-3-step-2
- step-6:divide the step4/step5

```python
visa_df=pd.read_csv("C:\\Users\\LENOVO\\OneDrive\\DATA SCIENCE\\data files\\Visadataset - Visadataset.csv")
```

```python

od=visa_df["prevailing_wage"]
min1=visa_df["prevailing_wage"].min()
max1=visa_df["prevailing_wage"].max()
NR=od-min1
DR=max1-min1
visa_df["prevailing_wage_min_max"]=NR/DR
visa_df[["prevailing_wage","prevailing_wage_min_max"]]
```
