```python
unique=visa_df["continent"].unique()
number=[]
for i in unique:
    con=visa_df["continent"]==i
    val=len(visa_df[con])
    number.append(val)
    print(f"{i}:{val}")
number
```

```python
continent_df=pd.DataFrame(zip(unique,number),columns=["Continent","Count"])
continent_df
```

```python
plt.bar("Continent","Count",data=continent_df)
plt.show()
```
